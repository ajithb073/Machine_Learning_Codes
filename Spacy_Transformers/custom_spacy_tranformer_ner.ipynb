```python
sentence_list = []
one_iter_entities = []
final_data = []
tempsentence_list = []
def sent_entity(sentences, words_list):
  tempsentence = sentences
  words_list = ast.literal_eval(words_list)
  # print(words_list)
  # print("*******")
  words_list = [item for sublist in words_list for item in sublist]
  # print(words_list)
  words_list = list(dict.fromkeys(words_list))
  # print("Old :", words_list)
  words_list= [[x] for x in words_list]
  entities_list = []
  # print("New :", words_list)
  for words in words_list: 
    # print(words)
    for tags in words: 
      tags = str(tags)
      try:
        total_length = len(tags)
        start_index = tempsentence.index(tags)
        # print(start_index)
        end_index = start_index + total_length
        dummy_word =[]
        for i in range(start_index,end_index):
            dummy_word.append('*')
        listToStr = ''.join([str(elem) for elem in dummy_word])
        tempsentence = tempsentence.replace(tags,listToStr)
        # print(tempsentence)
        output = ("[%d, %d, 'TECH']" %(start_index, end_index))
        output = ast.literal_eval(output)
        # print(output)
        entities_list.append(output)
      except:
        pass  
  sentence_list.append(sentences)
  print(sentences) 
  tempsentence_list.append(tempsentence)
  entities = {'entities' : entities_list}
  print(entities)
  final_format = [sentences,entities]
  one_iter_entities.append(entities)   
  return final_format
  

```

```python
TRAIN_DATA = annotation_data[0:80000]
TEST_DATA = annotation_data[80000:90984]
VALID_DATA = annotation_data[90984:]

print(len(TRAIN_DATA))
print(len(TEST_DATA))
print(len(VALID_DATA))
```

```python
import pandas as pd
from tqdm import tqdm
import spacy
from spacy.tokens import DocBin
nlp = spacy.blank("en") # load a new spacy model
db = DocBin() # create a DocBin object

```

```python
for text, annot in tqdm(TRAIN_DATA): # data in previous format
    doc = nlp.make_doc(text) # create doc object from text
    ents = []
    for start, end, label in annot["entities"]: # add character indexes
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    doc.ents = ents # label the text with the ents
    db.add(doc)

db.to_disk("./train.spacy") # save the docbin object

db = DocBin()
for text, annot in tqdm(VALID_DATA): # data in previous format
    doc = nlp.make_doc(text) # create doc object from text
    ents = []
    for start, end, label in annot["entities"]: # add character indexes
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    doc.ents = ents # label the text with the ents
    db.add(doc)

db.to_disk("./valid.spacy") # save the docbin object
```
